```python
from dotenv import load_dotenv
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_ollama import ChatOllama

load_dotenv()
```

### 1. Messages - The Building Blocks

```python
llm = ChatOllama(model="gemma3",
                 base_url="http://localhost:11434",
                 temperature=0.7,
                 num_predict=1024,
                 num_ctx=2048)
```

```python
llm
```

```python
msgs = [
    SystemMessage("You are a helpful assistant."),
    HumanMessage("Write a short poem on sea.")
]
msgs
```

```python
response =llm.invoke(msgs)
response
```

```python
response.pretty_print()
```

### 2. Prompt Templates - Reusable Prompts

```python
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "You are an expert in {subject}. Explain it to a {audience} level."),
    ("human", "Can you explain {topic}?")
])
prompt_template
```

```python
msgs = prompt_template.invoke({
    "subject": "Physics",
    "audience": "high school",
    "topic": "Quantum Mechanics"
})
msgs
```

```python
response = llm.invoke(msgs)
response
```

```python
response.pretty_print()
```

```python
msgs = prompt_template.invoke({
    "subject": "Physics",
    "audience": "Ph. D.",
    "topic": "Quantum Mechanics"
})
response = llm.invoke(msgs)
response.pretty_print()
```
